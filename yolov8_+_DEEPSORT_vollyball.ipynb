#**Demo Video 1**

```python
import cv2

def jpg_to_mp4(input_image_path, output_video_path, fps=30):
    # Read the input image
    image = cv2.imread(input_image_path)
    height, width, _ = image.shape

    # Create video writer object
    video_writer = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

    # Write the image to the video for a duration of 5 seconds
    duration_seconds = 5
    for _ in range(duration_seconds * fps):
        video_writer.write(image)

    # Release the video writer
    video_writer.release()

# Example usage
input_image_path = "/content/pv4.png"
output_video_path = "/content/output_video.mp4"
jpg_to_mp4(input_image_path, output_video_path, fps=5)
```
